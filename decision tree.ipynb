```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
```

# Decision Tree on diabetes Data Set

```python
Data = pd.read_csv('diabetes.csv')
InputTrain = Data.drop('Outcome',axis='columns')
InputTarget = Data['Outcome']
```

```python
Data.head()
```

```python
train_x, test_x, train_y, test_y =train_test_split(InputTrain,InputTarget,test_size = 0.3)

model = tree.DecisionTreeClassifier()
y =model.fit(train_x,train_y)
DT_pred = model.predict(test_x)

acc = accuracy_score(test_y, DT_pred)
print(acc)
```

```python
tree.plot_tree(y)
```

```python
import graphviz
dot_data = tree.export_graphviz(y, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("iris")
```
